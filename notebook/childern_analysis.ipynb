```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
children = pd.read_csv(
    "../data/Cyber Crime Against Children in 2020.csv"
)
```

```python
children.head()
children.shape
children.columns
children.info()
```

```python
children.isnull().sum()
```

```python
children = children.drop(
    columns=[
        'Unnamed: 8',
        'Unnamed: 9',
        'Unnamed: 10',
        'Unnamed: 11'
    ]
)
```

```python
children.columns
```

```python
top10_children = children.sort_values(
    by='Total Cyber crimes against Children',
    ascending=False
).head(10)

top10_children[
[
'State',
'Total Cyber crimes against Children'
]
]
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top10_children,
    x='Total Cyber crimes against Children',
    y='State'
)

plt.title(
    'Top States by Cyber Crimes Against Children'
)

plt.show()
```

```python
crime_types = children.drop(
    columns=[
        'State',
        'Total Cyber crimes against Children'
    ]
)

crime_totals = crime_types.sum()

crime_totals.sort_values(
    ascending=False
)
```

```python
plt.figure(figsize=(12,6))

crime_totals.sort_values(
    ascending=False
).plot(
    kind='bar'
)

plt.title(
    'Cyber Crime Categories Against Children'
)

plt.ylabel(
    'Number of Cases'
)

plt.show()
```

```python
crime_percentage = (
    crime_totals /
    crime_totals.sum()
)*100

crime_percentage.sort_values(
    ascending=False
)
```

```python
children['Children_Risk_Score'] = (
    children['Total Cyber crimes against Children']
    /
    children['Total Cyber crimes against Children'].max()
)*100
```

```python
children.sort_values(
    by='Children_Risk_Score',
    ascending=False
)[
[
'State',
'Children_Risk_Score'
]
].head(10)
```

```python
numeric_cols = children.select_dtypes(
    include=['int64']
)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title(
    'Children Cyber Crime Correlation Matrix'
)

plt.show()
```
