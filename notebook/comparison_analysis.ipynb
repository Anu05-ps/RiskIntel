```python
import pandas as pd
```

```python
women = pd.read_csv(
    "../data/Cyber Crime Against Women in 2020.csv"
)
```

```python
children = pd.read_csv(
    "../data/Cyber Crime Against Children in 2020.csv"
)
```

```python
children = children.drop(
    columns=[
        'Unnamed: 8',
        'Unnamed: 9',
        'Unnamed: 10',
        'Unnamed: 11'
    ]
)
```

```python
women_compare = women[
[
'State',
'Total Cyber Crimes against Women'
]
]

children_compare = children[
[
'State',
'Total Cyber crimes against Children'
]
]
```

```python
comparison = pd.merge(
    women_compare,
    children_compare,
    on='State'
)

comparison.head()
```

```python
comparison['Vulnerable_Population_Index'] = (
    comparison['Total Cyber Crimes against Women']
    +
    comparison['Total Cyber crimes against Children']
)
```

```python
comparison.sort_values(
    by='Vulnerable_Population_Index',
    ascending=False
).head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_states = comparison.sort_values(
    by='Vulnerable_Population_Index',
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top_states,
    x='Vulnerable_Population_Index',
    y='State'
)

plt.title(
    'Top States by Vulnerable Population Index'
)

plt.show()
```

```python
comparison.to_csv(
    "../dashboard/vulnerable_population.csv",
    index=False
)
```
